### Weather Data

```python
Airports=['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
```

```python
dd=pd.DataFrame()
```

```python
for i in Airports:
    path='C:/Users/SAI/OneDrive - University at Buffalo/2nd sem/DIC/weather/'+i
    for j in os.listdir(path):
        if('2016' in j):
            with open('C:/Users/SAI/OneDrive - University at Buffalo/2nd sem/DIC/weather/'+i+'/'+j) as f:
                d = json.load(f)
                data = json_normalize(d['data']['weather'])
                data=data[['date','hourly']]
                #data['airport']=i
                #works_data = json_normalize(data=d['data']['weather'], record_path='hourly', errors='ignore')
                works_data = json_normalize(data=d['data']['weather'], record_path=['hourly'], errors='ignore',meta=['date'])
                works_data['airport'] =i
                dd=dd.append(works_data)
```

```python
finalweatherdata2016=dd[['airport','date','windspeedKmph','winddirDegree','weatherCode','precipMM','visibility',
'pressure',
'cloudcover',
'DewPointF',
'WindGustKmph',
'tempF',
'WindChillF',
'humidity',
'time']]  
```

```python
dd2017=pd.DataFrame()
for i in Airports:
    path='C:/Users/SAI/OneDrive - University at Buffalo/2nd sem/DIC/weather/'+i
    for j in os.listdir(path):
        if('2017' in j):
            with open('C:/Users/SAI/OneDrive - University at Buffalo/2nd sem/DIC/weather/'+i+'/'+j) as f:
                d = json.load(f)
                data = json_normalize(d['data']['weather'])
                data=data[['date','hourly']]
                data['airport']=i
                works_data = json_normalize(data=d['data']['weather'], record_path='hourly', errors='ignore',meta=['date'])
                works_data['airport'] =i
                dd2017=dd2017.append(works_data)
```

```python
finalweatherdata2017=dd2017[['airport','date','windspeedKmph','winddirDegree','weatherCode','precipMM','visibility',
'pressure',
'cloudcover',
'DewPointF',
'WindGustKmph',
'tempF',
'WindChillF',
'humidity',
'time']]  
```
